```python
#Outlier Detection
def detect_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")
    return outliers
```

```python
detect_outliers('cost')
detect_outliers('rating')
detect_outliers('rating_count')
```

```python
df
```

```python
#Capping
Q1 = df['cost'].quantile(0.25)
Q3 = df['cost'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['cost'] = df['cost'].clip(lower, upper)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('/content/cleaned_swiggy (2).csv')
df
```

```python
# Create effective rating
df['effective_rating'] = df['rating'] * (df['rating_count'] / (df['rating_count'] + 50))
```

```python
# Keep display data separately
display_df = df[['name', 'city', 'cuisine', 'rating', 'cost']]
```

```python
display_df
```

```python
# Model data
model_df = df.drop(columns=['name'])
```

```python
model_df
```

```python
# ======================================
# ENCODING (CITY + CUISINE)
# ======================================
```

```python
categorical_cols = ['city', 'cuisine']

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

encoded_cat = encoder.fit_transform(model_df[categorical_cols])

encoded_cat_df = pd.DataFrame(
    encoded_cat,
    columns=encoder.get_feature_names_out(categorical_cols)
)
```

```python
encoded_cat_df
```

```python
# ======================================
# NUMERICAL FEATURES
# ======================================
```

```python
numerical_cols = ['effective_rating', 'rating_count', 'cost', 'no_rating']
```

```python
num_df = model_df[numerical_cols].reset_index(drop=True)
```

```python
num_df
```

```python
# ======================================
# SCALING
# ======================================
```

```python
scaler = StandardScaler()
scaled_num = scaler.fit_transform(num_df)
scaled_num_df = pd.DataFrame(scaled_num, columns=numerical_cols)
```

```python
scaled_num_df
```

```python
# ======================================
# FINAL DATASET
# ======================================
```

```python
final_df = pd.concat([scaled_num_df, encoded_cat_df], axis=1)

print("Final shape:", final_df.shape)
```

```python
# ======================================
# KMEANS CLUSTERING
# ======================================
```

```python
kmeans = KMeans(n_clusters=6, random_state=42)
df['cluster'] = kmeans.fit_predict(final_df)
```

```python
# Take sample (10,000 rows)
sample_df = final_df.sample(n=10000, random_state=42)

# Get labels for sample
sample_labels = df.loc[sample_df.index, 'cluster']

# Calculate score
score = silhouette_score(sample_df, sample_labels)

print("Silhouette Score (sample):", score)
```
